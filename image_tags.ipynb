```python
import os
import json
from PIL import Image
import torch
from transformers import pipeline
```

```python
image_folder = '/path.../images'
tag_list = ["mountains", "animals", "nature", "forest", "landscape", "flag", "map", "historic", "beach", "lighthouse", "ocean", "seaside", "wilderness", "waterfall", "sunset", "botanical", "garden", "flower", "tree", "castle", "bridge", "market", "main square", "fountain", "harbor", "train", "tram", "trolly", "street", "urban", "monument", "church", "boat", "car", "plane", "fashion", "wildlife", "illustration", "painting", "architectural", "plant", "slogan", "language", "sports", "love", "food", "religion", "skyline", "culture", "arts", "comic", "photograph", "drawing", "graphic", "writing", "greeting", "seasons", "landmark", "celebrity", "recipe", "poetry", "transportation", "people", "technology", "camping", "resort", "celebration", "holiday"]
threshold = 0.1

image_classifier = pipeline("zero-shot-image-classification", model="openai/clip-vit-base-patch32")

image_tag_output = []
valid_extensions = ('.jpg', '.jpeg', '.png')

for image_name in os.listdir(image_folder):
    if image_name.lower().endswith(valid_extensions):
        file_path = os.path.join(image_folder, image_name)

        try:
            img = Image.open(file_path).convert('RGB')
            results = image_classifier(img, candidate_labels=tag_list)
            assigned_tags = [
                res['label'] for res in results
                if res['score'] > threshold
            ]

            image_tag_output.append({
                "image_name": image_name,
                "tags": assigned_tags,
                "top_score": round(results[0]['score'], 4)
            })

        except Exception as e:
            print(f"Failed to process {image_name}: {e}")

with open('image_tags.json', 'w', encoding='utf-8') as f:
    json.dump(image_tag_output, f, indent=4)

print(f"Processed: {len(image_tag_output)}")
```
